```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans

%matplotlib inline
```

```python
data = pd.read_csv('./data/cluster_unix_time_indoor.csv')
data.columns = ['a', 'b', 'c', 'd']
data.head()
```

```python
X1, y1 = datasets.make_circles(n_samples=3000, factor=.6, noise=0.05)
X2, y2 = datasets.make_blobs(n_samples=2500, n_features=4, centers=[[1, 1.5]], cluster_std=[[.3]])
```

```python
X = np.concatenate((X1, X2))
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.show()
```

```python
y_pred = KMeans(n_clusters=3, random_state=9).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
y_pred = DBSCAN().fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
y_pred = DBSCAN().fit_predict(X[:, 0].reshape(-1, 1))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
y_pred = DBSCAN(eps=0.1).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
y_pred = DBSCAN(eps = 0.1, min_samples = 5).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```
